```python
import nltk
nltk.download("punkt")
```

```python
sentence1="My name is Melvin, I'm currently pursuing my masters degree in artificial intelligence"
sentence1=sentence1.lower()
sentence2="My name is Infant, I'm doing masters in artificial intelligence right now"
sentence2=sentence2.lower()
```

```python
from nltk.tokenize import word_tokenize
sent1=word_tokenize(sentence1)
sent2=word_tokenize(sentence2)
print(sent1)
print(sent2)
```

```python
def remove_punct(token):
  return [word for word in token if word.isalpha()]
sent1=remove_punct(sent1)
sent2=remove_punct(sent2)
print(sent1)
print(sent2)
```

LEMMATIZATION

```python
nltk.download("wordnet")
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
sent1 = [lemmatizer.lemmatize(words_sent) for words_sent in sent1]
sent2 = [lemmatizer.lemmatize(words_sent) for words_sent in sent2]
print(sent1)
print(sent2)
```

STOP WORDS

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
sent1=[word for word in sent1 if word not in stopwords.words('english')]
sent2=[word for word in sent2 if word not in stopwords.words('english')]
print(sent1)
print(sent2)
```

```python
import numpy as np
from scipy.spatial.distance import cosine
embeddings_index = {}
with open('glove.6B.100d.txt', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs
embeddings1 = [embeddings_index.get(word, np.zeros(100)) for word in sent1]
embeddings2 = [embeddings_index.get(word, np.zeros(100)) for word in sent2]
sentence_embedding1 = np.mean(embeddings1, axis=0)
sentence_embedding2 = np.mean(embeddings2, axis=0)
similarity = 1 - cosine(sentence_embedding1, sentence_embedding2)
print('Cosine similarity:', similarity)
```
